```python
pricedf=df[['ID','price']]
pricedf.head(3)
```

```python
conditions = [
    (pricedf['price'] <= 10),
    (pricedf['price'] > 10) & (pricedf['price'] <= 20),
    (pricedf['price'] > 20) & (pricedf['price'] <= 30),
    (pricedf['price'] > 30) & (pricedf['price'] <= 40),
    (pricedf['price'] > 40) & (pricedf['price'] <= 50),
    (pricedf['price'] > 50) & (pricedf['price'] <= 60),
    (pricedf['price'] > 60) & (pricedf['price'] <= 70),
    (pricedf['price'] > 70) & (pricedf['price'] <= 80),
    (pricedf['price'] > 80) & (pricedf['price'] <= 90),
    (pricedf['price'] > 90) & (pricedf['price'] <= 100),
    (pricedf['price'] > 100) & (pricedf['price'] <= 110),
    (pricedf['price'] > 110) & (pricedf['price'] <= 120),
    (pricedf['price'] > 120) & (pricedf['price'] <= 130),
    (pricedf['price'] > 130) & (pricedf['price'] <= 140),
    (pricedf['price'] > 140)
    ]

# create a list of the values we want to assign for each condition
values = ['0-10','11-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','110-120','121-130','131-140','>140']

# create a new column and use np.select to assign values to it using our lists as arguments
pricedf['pricebucket'] = np.select(conditions, values)
```
